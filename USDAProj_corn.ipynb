```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
warnings.simplefilter('ignore', np.RankWarning)
```

```python
df = pd.read_csv("USDAProj_Corn_2007to2008.csv")
df
```

```python
df = df.iloc[1:24,23:29]
df
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(11, 8.5)
sns.regplot(data=df, 
            x="United States Imports", y="United States Exports", 
            order=2).set_title("United States Exports Vs imports")
```

```python
# increase fontsize and figsize
plt.figure(figsize=(6,6))
subject_names = ["United States Production","United States DomesticFeed", 
                 "United States DomesticTotal"]
subject_marks = [331, 150, 266]

colors = ["#ffd50544", "#7952b399", "#ff222244"]

plt.pie(subject_marks, labels = subject_names, autopct='%1.1f%%', startangle=90, 
        colors=colors, textprops={'fontsize': 14})
plt.axis('equal') 
plt.show()
```

```python
df[["United States DomesticFeed",
    "United States DomesticTotal", 
    "United States Exports"]].plot.barh(stacked=True, 
                                        figsize=(12, 12)).set_title("USDA Corn production 2007 to 2008")
```
